```python
def possible_positions_func(number):
    position=number[0]
    row=position[0]
    column=position[1]
    
    value=number[1]
    length=len(str(value))
    
    suspect_position=[]
    
    for suspect_row in (row-1,row+1):
        for suspect_column in range(column-1, column+length+1):
            suspect_position.extend([((suspect_row, suspect_column))])
    suspect_position.extend([((row, column-1)),((row, column+length))])
    return suspect_position
    
```

```python
import re

def extract_numbers_and_symbols(text, row=0, numbers=[], symbols={}):
    numbers.extend([((row, m.start()), m.group()) for m in re.finditer(r'\d+', text)])
    symbols.update({(row, m.start()): m.group() for m in re.finditer(r'\*', text)})
    
    
    return numbers, symbols

strings_list = input_strings.split('\n')
numbers, symbols = [], {}
for i in range(len(strings_list)):
    numbers, symbols = extract_numbers_and_symbols(strings_list[i], i, numbers, symbols)
print(numbers)
print(symbols)
```

```python
dict = { (a, b): 0 for a in range(len(strings_list)) for b in range(len(strings_list)) }
for number in numbers:
    positions = possible_positions_func(number)  # Assuming possible_positions_func is your function
    for i in positions:
        if i in dict:
            dict[i]+=1
            

possible_overlaps=[key for key, value in dict.items() if value == 2]
crucial_stars={}

for overlap in possible_overlaps:
    if overlap in symbols:
        crucial_stars.update({overlap: symbols[overlap]})
```
